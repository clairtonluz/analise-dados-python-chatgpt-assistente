```python
url = 'https://raw.githubusercontent.com/alura-cursos/python-analise-chatgpt-assistente/main/Dados/vendas_mensal_franquias.json'
```

```python
import pandas as pd

df_original = pd.read_json(url)

# Exibe as primeiras linhas do DataFrame
df_original.head()
```

```python
# Normalizando as colunas 'vendas' e 'loja'
df_vendas = pd.json_normalize(df_original['vendas'])
df_loja = pd.json_normalize(df_original['loja'])
# Concatenando as colunas normalizadas de volta ao DataFrame original
df = pd.concat([df_original.drop(columns=['vendas', 'loja']), df_vendas, df_loja], axis=1)

df.head()
```
